#### 3) Monte um gráfico de barras mostrando o APGAR5 médio por CONSULTAS

```python
plt.close('all')
plt.rc('figure', figsize=(10, 5))
sns.barplot(x='CONSULTAS', y='APGAR5', data=df, errorbar=None);
```

####  4) Monte um painel com 4 gráficos mostrando as frequencias de SEXO, GESTACAO, ESTCIVMAE e CONSULTAS

```python
plt.close('all')
plt.rc('figure', figsize=(12, 8))
fig, axes = plt.subplots(2, 2)


sns.countplot(ax = axes[0, 0], x='SEXO', data=df)
sns.countplot(ax = axes[0, 1], x='GESTACAO', data=df)
sns.countplot(ax = axes[0, 1], x='GESTACAO', data=df).set_xticklabels([
    '37 a 41 semanas',
    '42 semanas e mais',
    '32 a 36 semanas',
    '22 a 27 semanas',
    '28 a 31 semanas',
    'Menos de 22 semanas'],
    rotation=30)
sns.countplot(ax = axes[1, 0], x='ESTCIVMAE', data=df).set_xticklabels([
    'solteira',
    'casada',
    'viúva',
    'divorciado',
    'união consensual',
    'ignorado'],
    rotation=30)
sns.countplot(ax = axes[1, 1], x='CONSULTAS', data=df)
plt.subplots_adjust(wspace=0.25,hspace=0.75)
```
